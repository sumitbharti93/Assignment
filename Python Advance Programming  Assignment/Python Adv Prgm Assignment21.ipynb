```python
'''
1. Given a sentence, return the number of words which have the same first and last letter.

Examples

count_same_ends("Pop! goes the balloon") ➞ 1

count_same_ends("And the crowd goes wild!") ➞ 0

count_same_ends("No I am not in a gang.") ➞ 1
'''
#Code
def count_same_ends(st):
    lst = st.split(' ')
    count=0
    for i in range(len(lst)):
        if len(lst[i])>=2:
            if not 97<=ord(lst[i][-1])<=122:
                if lst[i][0].lower()==lst[i][-2].lower():
                    count+=1
            else:
                if lst[i][0].lower()==lst[i][-1].lower():
                    count+=1
    return count

if __name__ == "__main__":
    st = input("Enter the string ")
    result = count_same_ends(st)
    print(result)
```

```python
'''
2. The Atbash cipher is an encryption method in which each letter of a word is replaced with 
its "mirror" letter in the alphabet: A <=> Z; B <=> Y; C <=> X; etc.

Create a function that takes a string and applies the Atbash cipher to it.

Examples

atbash("apple") ➞ "zkkov"

atbash("Hello world!") ➞ "Svool dliow!"

atbash("Christmas is the 25th of December") ➞ "Xsirhgnzh rh gsv 25gs lu Wvxvnyvi"
'''
def atbash(st):
    atbash_word = []
    lst_sentence = st.split(" ")
    
    for i in range(len(lst_sentence)):
        lst_word = list(lst_sentence[i])
        
        for j in range(len(lst_word)):
            
            if 97<=ord(lst_word[j])<=122:
                a = ord(lst_word[j])
                b = a-97
                c = 122-b
                lst_word[j]= chr(c)
                
            elif 65<=ord(lst_word[j])<=90:
                
                a = ord(lst_word[j])
                b = a-65
                c = 90-b
                lst_word[j]= chr(c)
                
        word = "".join(lst_word)
        atbash_word.append(word)
    atbash_word = " ".join(atbash_word)
    
    return atbash_word

if __name__ == "__main__":
    st = input("Enter the string ")
    result = atbash(st)
    print(result)
```

```python
'''
3. Create a class Employee that will take a full name as argument, as well as a set of none, 
one or more keywords. Each instance should have a name and a lastname attributes plus one more 
attribute for each of the keywords, if any.

Examples

john = Employee("John Doe")
mary = Employee("Mary Major", salary=120000)
richard = Employee("Richard Roe", salary=110000, height=178)
giancarlo = Employee("Giancarlo Rossi", salary=115000, height=182, nationality="Italian")

john.name ➞ "John"
mary.lastname ➞ "Major"
richard.height ➞ 178
giancarlo.nationality ➞ "Italian"
'''
#Code
class Employee:
    def __init__(self, full_name, salary = None, height= None, nationality = None):
        self.name = full_name.split(" ")[0]
        self.last_name = full_name.split(" ")[1]
        self.salary = salary
        self.height = height
        self.nationality = nationality
        
john = Employee("John Doe")
print(john.name)
mary = Employee("Mary Major", salary=120000)
print(mary.last_name)
richard = Employee("Richard Roe", salary=110000, height=178)
print(richard.height)
giancarlo = Employee("Giancarlo Rossi", salary=115000, height=182, nationality="Italian")
print(giancarlo.nationality)
```

```python
'''
4. Create a function that determines whether each seat can "see" the front-stage. A number can 
"see" the front-stage if it is strictly greater than the number before it.

Everyone can see the front-stage in the example below:

# FRONT STAGE
[[1, 2, 3, 2, 1, 1],
[2, 4, 4, 3, 2, 2],
[5, 5, 5, 5, 4, 4],
[6, 6, 7, 6, 5, 5]]

# Starting from the left, the 6 > 5 > 2 > 1, so all numbers can see.
# 6 > 5 > 4 > 2 - so all numbers can see, etc.

Not everyone can see the front-stage in the example below:

# FRONT STAGE
[[1, 2, 3, 2, 1, 1], 
[2, 4, 4, 3, 2, 2], 
[5, 5, 5, 10, 4, 4], 
[6, 6, 7, 6, 5, 5]]

# The 10 is directly in front of the 6 and blocking its view.

The function should return True if every number can see the front-stage, and False if even a 
single number cannot.

Examples

can_see_stage([
  [1, 2, 3],
  [4, 5, 6],
  [7, 8, 9]
]) ➞ True

can_see_stage([
  [0, 0, 0],
  [1, 1, 1],
  [2, 2, 2]
]) ➞ True

can_see_stage([
  [2, 0, 0], 
  [1, 1, 1], 
  [2, 2, 2]
]) ➞ False

can_see_stage([
  [1, 0, 0],
  [1, 1, 1],
  [2, 2, 2]
]) ➞ False

# Number must be strictly smaller than 
# the number directly behind it.
'''
#Code
def can_see_stage(lst):
    flag=0
    for column in range(len(lst[0])):
        for row in range(len(lst)-1):
            if lst[row][column]>lst[row+1][column]:
                flag =1
                break
    if flag==0:
        return True
    else:
        return False
    
if __name__ == "__main__":
    lst = eval(input("Enter the list "))
    result = can_see_stage(lst)
    print(result)
```

```python
'''
5. Create a Pizza class with the attributes order_number and ingredients (which is given as a 
list). Only the ingredients will be given as input.

You should also make it so that its possible to choose a ready made pizza flavour rather than 
typing out the ingredients manually! As well as creating this Pizza class, hard-code the 
following pizza flavours.

Name	                    Ingredients
hawaiian	          ham, pineapple
meat_festival	beef, meatball, bacon
garden_feast	spinach, olives, mushroom

Examples

p1 = Pizza(["bacon", "parmesan", "ham"])    # order 1
p2 = Pizza.garden_feast()                  # order 2

p1.ingredients ➞ ["bacon", "parmesan", "ham"]

p2.ingredients ➞ ["spinach", "olives", "mushroom"]

p1.order_number ➞ 1

p2.order_number ➞ 2
'''
#Code
class Pizza:
    count=0
    def __init__(self, ingrediants=None):
        self.ingrediants = ingrediants
        Pizza.count+=1
        self.order_number = Pizza.count
        
    @staticmethod
    def hawaiian(self):
        Pizza.count+=1
        return ['ham', 'pineapple']
    @staticmethod
    def meat_festival(self):
        return ["beef", "meatball", "bacon"]
    @staticmethod
    def garden_feast():
        return ["spinach", "olives", "mushroom"]
    
p1 = Pizza(["bacon", "parmesan", "ham"])
print(p1.ingrediants)
print(p1.order_number)
p2 = Pizza(Pizza.garden_feast())
print(p2.ingrediants)
print(p2.order_number)
```
