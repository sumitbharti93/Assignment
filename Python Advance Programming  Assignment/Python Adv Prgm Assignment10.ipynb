```python
'''
1. Create a function that takes the width, height and character and returns a picture frame as a 2D list.

Examples

get_frame(4, 5, "#") ➞ [
  ["####"],
  ["#  #"],
  ["#  #"],
  ["#  #"],
  ["####"]
]
# Frame is 4 characters wide and 5 characters tall.


get_frame(10, 3, "*") ➞ [
  ["**********"],
  ["*        *"],
  ["**********"]
]
# Frame is 10 characters and wide and 3 characters tall.


get_frame(2, 5, "0") ➞ "invalid"
# Frame's width is not more than 2.

'''
#Code
def get_frame(width, height, char):
    frame = []
    if width >=2:
        for i in range(height):
            l =""
            if i==0 or i == height-1:
                for j in range(width):
                    l = l+char
                frame.append([l])
            else:
                for j in range(width):
                    if j==0 or j==width-1:
                        l = l+char
                    else:
                        l+=" "
                        
                frame.append([l]) 
    else:
        return "Frame width is not more than 2 "
    
    return frame

if __name__ == "__main__":
    width, height,char = eval(input("Enter the widht , height and char "))
    result = get_frame(width, height, char)
    print(result)

```

```python
'''
2. Write three functions:

  1. boolean_and
  2. boolean_or
  3. boolean_xor
These functions should evaluate a list of True and False values, starting from the leftmost element and evaluating 
pairwise.

Examples

boolean_and([True, True, False, True]) ➞ False
# [True, True, False, True] => [True, False, True] => [False, True] => False

boolean_or([True, True, False, False]) ➞ True
# [True, True, False, True] => [True, False, False] => [True, False] => True

boolean_xor([True, True, False, False]) ➞ False
# [True, True, False, False] => [False, False, False] => [False, False] => False
'''
#Code

def boolean_and(lst):
    a = True
    for i in range(len(lst)):
        a =  a and lst[i]
    return a 
    
def boolean_or(lst):
    a = False 
    for i in range(len(lst)):
        a = a or lst[i]
    return a 
    
def boolean_xor(lst):
    a = False
    for i in range(len(lst)):
        a = a ^ lst[i]
    return a 

if __name__ == "__main__":
    lst1 = eval(input("Enter the list for and function "))
    result = boolean_and(lst1)
    print(result)
    
    lst2 = eval(input("Enter the list for or function "))
    result = boolean_or(lst2)
    print(result)
    
    
    lst3 = eval(input("Enter the list for xor function "))
    result = boolean_xor(lst3)
    print(result)

```

```python
'''
3. Create a function that creates a box based on dimension n.

Examples

make_box(5) ➞ [
  "#####",
  "#   #",
  "#   #",
  "#   #",
  "#####"
]

make_box(3) ➞ [
  "###",
  "# #",
  "###"
]

make_box(2) ➞ [
  "##",
  "##"
]

make_box(1) ➞ [
  "#"
]
'''
def make_box(num):
    box = []
    char = '#'
    for i in range(num):
        l =""
        if i==0 or i == num-1:
            for j in range(num):
                l = l+char
            box.append(l)
        else:
            for j in range(num):
                if j==0 or j==num-1:
                    l = l+char
                else:
                    l+=" "
            box.append(l) 
    return box

if __name__ == "__main__":
    num = int(input("Enter the number  "))
    result = make_box(num)
    print(result)
    
    
```

```python
'''
4. Given a common phrase, return False if any individual word in the phrase contains duplicate letters. Return True 
otherwise.

Examples

no_duplicate_letters("Fortune favours the bold.") ➞ True

no_duplicate_letters("You can lead a horse to water, but you can't make him drink.") ➞ True

no_duplicate_letters("Look before you leap.") ➞ False
# Duplicate letters in "Look" and "before".

no_duplicate_letters("An apple a day keeps the doctor away.") ➞ False
# Duplicate letters in "apple", "keeps", "doctor", and "away".
'''
#Code

def no_duplicate_letters(st):
    lst_st = st.split(" ")
    count=0
    for i in range(len(lst_st)):
        flag=0
        for j in range(len(lst_st[i])):
            if lst_st[i].count(lst_st[i][j])==2:
                flag=1
                
        if flag==0:
            count+=1
        else:
            return False
    if count==len(lst_st):
        return True 


if __name__ == "__main__":
    st = input("Enter the string  ")
    result = no_duplicate_letters(st)
    print(result)
    

```

```python
'''

5. Write a regular expression that will match the states that voted yes to President Trump's impeachment. You must use
RegEx positive lookahead.

Example

txt = "Texas = no, California = yes, Florida = yes, Michigan = no"
pattern = "yourregularexpressionhere"

re.findall(pattern, txt) ➞ ["California", "Florida"]
'''
#Code
import re 
def func1(st):
    pattern = re.compile(r'\w+(?= = yes)')
    match = re.findall(pattern, st)
    return match

if __name__ == "__main__":
    st = input("Enter the string  ")
    result = func1(st)
    print(result)

```
