```python
'''
1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more 
letter at a random position. Return the letter that was added to t.

Examples

find_the_difference("abcd", "abcde") ➞ "e"

find_the_difference("", "y") ➞ "y"

find_the_difference("ae", "aea") ➞ "a"

'''
#Code

def find_the_difference(s,t):
    ord_s_total=0
    ord_t_total = 0
    for i in range(len(s)):
        ord_s_total+=ord(s[i])
        ord_t_total+=ord(t[i])
        
    ord_t_total+=ord(t[-1])
    diff = ord_t_total-ord_s_total
    extra_letter = chr(diff)
    return extra_letter
        
        
if __name__ =="__main__":
    s,t  = eval(input("Enter the two strings "))
    result = find_the_difference(s,t)
    print(result)
```

```python
'''
2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. 
Finally return the total in a list.

List order is:

[int, str, bool, list, tuple, dictionary]

Examples

count_datatypes(1, 45, "Hi", False) ➞ [2, 1, 1, 0, 0, 0]

count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) ➞ [3, 0, 0, 1, 1, 0]

count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23) ➞ 
[2, 2, 3, 1, 0, 2]

count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]) ➞ [2, 0, 1, 2, 2, 0]
'''

def count_datatypes(lst):
    int_count = 0
    str_count=0
    bool_count=0
    list_count =0
    tuple_count = 0
    dict_count = 0
    for i in range(len(lst)):
        if type(lst[i])==int:
            int_count+=1
        elif type(lst[i])==str:
            str_count+=1
        elif type(lst[i])==bool:
            bool_count+=1
        elif type(lst[i])==list:
            list_count+=1
        elif type(lst[i])==tuple:
            tuple_count+=1
        elif type(lst[i])==dict:
            dict_count+=1
    type_count = [int_count,str_count,bool_count,list_count,tuple_count,dict_count]
    return type_count
        
if __name__ =="__main__":
    lst = eval(input("Enter the lists "))
    result = count_datatypes(lst)
    print(result)
```

```python
'''
3. A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English 
alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together 
as it progresses in a similar fashion as the Fibonacci series.

Examples

fib_str(3, ["j", "h"]) ➞ "j, h, hj"

fib_str(5, ["e", "a"]) ➞ "e, a, ae, aea, aeaae"

fib_str(6, ["n", "k"]) ➞ "n, k, kn, knk, knkkn, knkknknk"

'''
#Code
def fib_str(num, lst):
    a = lst[0]
    b = lst[1]
    fib_st = ""
    for i in range(num):
        fib_st+=a
        c =b+a
        a =b
        b = c
        fib_st+= ", "
    return fib_st

if __name__ =="__main__":
    num, lst = eval(input("Enter the lists "))
    result = fib_str(num, lst)
    print(result)

```

```python
'''
4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this
format:

"nines:your answer, threes:your answer, ones:your answer"

You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given 
integer when multiplied by one, three or nine (depends).

Examples

ones_threes_nines(10) ➞ "nines:1, threes:0, ones:1"

ones_threes_nines(15) ➞ "nines:1, threes:2, ones:0"

ones_threes_nines(22) ➞ "nines:2, threes:1, ones:1"

'''
#Code

def ones_threes_nines(num):
    ones_count = 0
    threes_count=0
    nines_count = 0
    if num>0 and num<3:
        ones_count+=num
    elif num>=3 and num<9:
            threes_count+= num//3
            ones_count+=num%3
    elif num>=9 and num<=26:
        nines_count+=num//9
        a = num%9
        threes_count+=a//3
        ones_count+=a%3
    else:
        return 'number out of range, input allowed between 1 and 26'
    st = "nines:"+str(nines_count)+", "+"thress:" + str(threes_count)+ ", " + "ones:" + str(ones_count)
    return st


if __name__ =="__main__":
    num = int(input("Enter the number "))
    result = ones_threes_nines(num)
    print(result)
```

```python
'''
5. The Fibonacci sequence is a classic use case for recursive functions since the value of the sequence at a given 
index is dependent on the sum of the last two values. However, the recursion tree created by solving the Fibonacci 
sequence recursively can grow quite fast. Therefore it can be important to think about the implications of running a 
function recursively. Depending on the size of n needed and the capabilities of the system in question you might want to 
take a different approach.

Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n.

Examples

fib(6) ➞ 8
# 0 + 1 = 1, 1 + 1 = 2, 1 + 2 = 3, 2 + 3 = 5, 3 + 5 = 8

fib(1) ➞ 1

fib(2) ➞ 1

'''
def fib(n):
    a = 0
    b = 1
    for i in range(n):
        c = a+b
        a =b
        b =c
        
    return a

if __name__ =="__main__":
    num = int(input("Enter the number "))
    result = fib(num)
    print(result)



```
