```python
# 1. Write a Python program to find sum of elements in list?
# Code

def list_sum(l):
    try:
        sum_elements = 0
        for i in range(len(l)):
            sum_elements+=l[i]
        return sum_elements
    except Exception as e:
        print(e)
def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        print("Enter the elements")
        for i in range(num_elements):
            l.append(int(input()))
        return l
    except Exception as e:
        print(e)

def get_result():
    try:
        l = get_input()
        result = list_sum(l)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
    result = get_result()
    print('Sum of the elemets of the given list is ', result)



```

```python
# 2. Write a Python program to  Multiply all numbers in the list?
#Code

def list_mul(l):
    try:
        mul_elements = 1
        for i in range(len(l)):
            mul_elements*=l[i]
        return mul_elements
    except Exception as e:
        print(e)
def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        print("Enter the elements")
        for i in range(num_elements):
            l.append(int(input()))
        return l
    except Exception as e:
        print(e)

def get_result():
    try:
        l = get_input()
        result = list_mul(l)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
    result = get_result()
    print('Multiplication of the elemets of the given list is ', result)


```

```python
# 3. Write a Python program to find smallest number in a list?
#Code

def list_smallest_number(l):
    try:
        smallest_number = l[0]
        for i in range(len(l)):
            if smallest_number>l[i]:
                smallest_number=l[i]
        return smallest_number
    except Exception as e:
        print(e)
def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        print("Enter the elements")
        for i in range(num_elements):
            l.append(int(input()))
        return l
    except Exception as e:
        print(e)

def get_result():
    try:
        l = get_input()
        result = list_smallest_number(l)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
    result = get_result()
    print('Smallest number of all the elemets in the given list is ', result)
```

```python
# 5. Write a Python program to find second largest number in a list?
#Code

def list_second_largest_number(l):
    try:
        first_largest = l[0]
        second_largest = 0
        if len(l)==1:
            return 1
        else:
            for i in range(1,len(l)):
                if first_largest<l[i]:
                    second_largest = first_largest
                    first_largest=l[i]
                elif second_largest<l[i]:
                        second_largest = l[i]
        return second_largest
    except Exception as e:
        print(e)
def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        print("Enter the elements")
        for i in range(num_elements):
            l.append(int(input()))
        return l
    except Exception as e:
        print(e)

def get_result():
    try:
        l = get_input()
        result = list_second_largest_number(l)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
    result = get_result()
    if result==1:
        print("invalid input, please enter element more than 1")
    else:
        print('Second Largest number of all the elemets in the given list is ', result)

```

```python
# 6. Write a Python program to find N largest elements from a list?
#Code


def list_N_largest_elements(l,N):
    try:
        l.sort()
        N_largest_element = l[len(l)-N:]
    
        return N_largest_element
    except Exception as e:
        print(e)
def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        print("Enter the elements")
        for i in range(num_elements):
            l.append(int(input()))
        N = int(input("Enter the number of N largest Elements "))
        return l,N
    except Exception as e:
        print(e)

def get_result():
    try:
        l,N = get_input()
        result = list_N_largest_elements(l,N)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
    result = get_result()
    print('N Largest elements of all the elemets in the given list is ', result)
```

```python
# 7. Write a Python program to print even numbers in a list?
#Code

def list_even_number(l):
    try:
        list_even = []
        for i in range(len(l)):
            if (l[i]%2)==0:
                list_even.append(l[i])
        return list_even
    except Exception as e:
        print(e)
def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        print("Enter the elements")
        for i in range(num_elements):
            l.append(int(input()))
        return l
    except Exception as e:
        print(e)

def get_result():
    try:
        l = get_input()
        result = list_even_number(l)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
    result = get_result()
    print('Even number of all the elemets in the given list is ', result)
    
```

```python
# 9. Write a Python program to Remove empty List from List?
#Code


def list_remove_empty_lst(l):
    try:
        final_list = []
        for i in range(len(l)):
            if (l[i])!=[]:
                final_list.append(l[i])
        return final_list
    except Exception as e:
        print(e)
def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        
        for i in range(num_elements):
            type_elements = input("Enter the type of elements, only 'int' and 'list' elments are allowed ")
            if type_elements == 'int':
                print("Enter the integer element ")
                l.append(int(input()))
            elif type_elements== 'list':
                number_elements = int(input("Enter the number of elemnts in given list "))
                b =[]
                for i in range(number_elements):
                    b.append(int(input()))
                l.append(b)
        return l
    except Exception as e:
        print(e)

def get_result():
    try:
        l = get_input()
        print("Final input list is ", l)
        result = list_remove_empty_lst(l)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
        result = get_result()
        print('Final list after removing empty list from given list is ', result)

    
```

```python
# 10. Write a Python program to Cloning or Copying a list?
#Code

def fun_list_clone(l):
    list_copy=[]
    for i in range(len(l)):
        list_copy.append(l[i])
    return list_copy

def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        print("Enter the elements")
        for i in range(num_elements):
            l.append(int(input()))
        return l
    except Exception as e:
        print(e)

def get_result():
    try:
        l = get_input()
        print("Final input list is ", l)
        result = fun_list_clone(l)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
    result = get_result()
    print('Final list after cloning ', result)
    
    
```

```python
# 11. Write a Python program to Count occurrences of an element in a list?
#Code

def count_occurences_list(l,element):
    try:
        count = 0
        for i in range(len(l)):
            if element==l[i]:
                count+=1
        return count,element
    except Exception as e:
        print(e)
def get_input():
    try:
        l=[]
        num_elements = int(input("Enter the number of elements in given list"))
        print("Enter the elements")
        for i in range(num_elements):
            l.append(int(input()))
        element = int(input("Enter the element for which count has to be determined "))
        return l,element
    except Exception as e:
        print(e)

def get_result():
    try:
        l,element = get_input()
        print('Final input list is ', l)
        result = count_occurences_list(l,element)
        return result
    except Exception as e:
        print(e)

if __name__ == "__main__" :
    result,element = get_result()
    print(f'Occurences of {element} in given list is ', result)
    


```
