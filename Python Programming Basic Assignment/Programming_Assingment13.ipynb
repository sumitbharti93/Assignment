```python
'''Question 1:

Write a program that calculates and prints the value according to the given formula:

Q = Square root of [(2 * C * D)/H]

Following are the fixed values of C and H:

C is 50. H is 30.

D is the variable whose values should be input to your program in a comma-separated sequence.

Example

Let us assume the following comma separated input sequence is given to the program:

100,150,180

The output of the program should be:

18,22,24'''


#Code
from math import sqrt

def fun_math_calculate(D_input):
    try:
        C =50
        H = 30
        final_sequence = []
        for D in D_input:
            Q = sqrt((2 * C * D)/H)
            final_sequence.append(int(Q))
        return final_sequence
    except Exception as e:
        print(e)

def get_input():
    print("enter the number of element in the sequence")
    num_elements = int(input())
    D_input = []
    for i in range(num_elements):
        D_input.append(int(input()))
        
    return D_input

def get_result():
    D_input = get_input()
    print("Given input sequence is ", D_input)
    result = fun_math_calculate(D_input)
    return result
    
if __name__ == "__main__":
    result = get_result()
    print("Sequence after computing given mathematical operation is ", result)
```

```python
''' 
Question 2:
Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The element value in the i-th 
row and j-th column of the array should be i*j.

Note: i=0,1.., X-1; j=0,1,¡¬Y-1.

Example

Suppose the following inputs are given to the program:

3,5

Then, the output of the program should be:

[[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]] 

'''


def fun_arr_generation(X,Y):
    final_matrix = [[0 for i in range(Y)] for i in range(X)]  # for storing matrix initializing matrix with zero as element
    for i in range(X):
        for j in range(Y):
            final_matrix[i][j] = i*j
    return final_matrix

def get_input():
    print("enter X and Y, matrix will be created with X rows and Y columns ")
    X = int(input())
    Y = int(input())
    
    return X, Y

def get_result():
    X, Y= get_input()
    print("Given input is ", X, Y)
    result = fun_arr_generation(X,Y)
    return result
    
if __name__ == "__main__":
    result = get_result()
    print("Final matrix is ", result)
```

```python
'''
Question 3:
Write a program that accepts a comma separated sequence of words as input and prints the words in a comma-separated sequence
after sorting them alphabetically.

Suppose the following input is supplied to the program:

without,hello,bag,world

Then, the output should be:

bag,hello,without,world

'''


def fun_words_sequence_sorting(words):
    for i in range(len(words)):
        for j in range(len(words)-1-i):
            if words[j]>words[j+1]:
                temp = words[j+1]
                words[j+1] = words[j]
                words[j] = temp
    return words

def get_input():
    num_words = int(input("Enter the number of words"))
    words = []
    for i in range(num_words):
        words.append(input())
        
    return words

def get_result():
    words = get_input()
    print("Given input is ", words )
    result = fun_words_sequence_sorting(words)
    return result
    
if __name__ == "__main__":
    result = get_result()
    print("Final sequence after sorting is ", result)

```

```python
"""
Question 4:
Write a program that accepts a sequence of whitespace separated words as input and prints the words after removing all 
duplicate words and sorting them alphanumerically.

Suppose the following input is supplied to the program:

hello world and practice makes perfect and hello world again

Then, the output should be:

again and hello makes perfect practice world

"""


def words_sequence_sorting2(words):
    words = words.split()
    words = list(set(words))
    for i in range(len(words)):
        for j in range(len(words)-1-i):
            if words[j]>words[j+1]:
                temp = words[j+1]
                words[j+1] = words[j]
                words[j] = temp
                
    words = " ".join(words)
    return words

def get_input():
    words = input()
    return words

def get_result():
    words = get_input()
    print("Given input is \n ", words)
    result = words_sequence_sorting2(words)
    return result
    
if __name__ == "__main__":
    result = get_result()
    print("Final sequence after sorting and removing duplicate element is \n", result)
    
```

```python
"""
Question 5:
Write a program that accepts a sentence and calculate the number of letters and digits.

Suppose the following input is supplied to the program:

hello world! 123

Then, the output should be:

LETTERS 10

DIGITS 3
"""

def sentence_letters_digits_calculate(sentence):
    sentence =sentence
    count_letter=0
    count_digit = 0
    for i in range(len(sentence)):
        if sentence[i].isalpha():
            count_letter+=1
        elif sentence[i].isdigit():
            count_digit+=1
    return count_letter,count_digit

def get_input():
    sentence = input()
    return sentence

def get_result():
    sentence = get_input()
    print("Given input is \n ", sentence)
    result = sentence_letters_digits_calculate(sentence)
    return result
    
if __name__ == "__main__":
    result = get_result()
    count_letter,count_digit = result
    print("In the given sentence, no of letter", count_letter)
    print("In the given sentence, no of digit", count_digit)
    
```

```python
'''
Question 6:
A website requires the users to input username and password to register. Write a program to check the validity of 
password input by users.

Following are the criteria for checking the password:

1. At least 1 letter between [a-z]

2. At least 1 number between [0-9]

1. At least 1 letter between [A-Z]

3. At least 1 character from [$#@]

4. Minimum length of transaction password: 6

5. Maximum length of transaction password: 12

Your program should accept a sequence of comma separated passwords and will check them according to the above criteria. 
Passwords that match the criteria are to be printed, each separated by a comma.

Example

If the following passwords are given as input to the program:

ABd1234@1,a F1#,2w3E*,2We3345

Then, the output of the program should be:

ABd1234@1

'''

#Code:

import re

def password_check(input_password):
    list_of_accepted_password = []
    for i in range(len(input_password)):
        count=0
        pattern = re.compile(r'[a-z]')
        if pattern.search(input_password[i]):
            count+=1
        pattern = re.compile(r'[A-Z]')
        if pattern.search(input_password[i]):
            count+=1
        pattern = re.compile(r'[0-9]')
        if pattern.search(input_password[i]):
            count+=1
        pattern = re.compile(r'[$#@]')
        if pattern.search(input_password[i]):
            count+=1
        if 6<=len(input_password[i])<=12:
            count+=1
        if count==5:
            list_of_accepted_password.append(input_password[i])        
        else:
            continue
    return list_of_accepted_password
def get_input():
    input_password = ['ABd1234@1','a F1#' ,'2w3E*','2We3345']
    return input_password
def get_result():
    input_password = get_input()
    print("List of Given password is \n ", input_password)
    result = password_check(input_password)
    return result
    
if __name__ == "__main__":
    result = get_result()
    print("List of accepted password ", result)
    
```
